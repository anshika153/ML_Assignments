```python
# Q1. Dataset Overview

from google.colab import files
uploaded = files.upload()

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("agriculture_yield_dataset.csv")

print("Rows and Columns:", df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nFirst 10 Records:")
display(df.head(10))
```

```python
# Q2. Data Types and Missing Values

print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Q3. Descriptive Statistics

display(df.describe())

numeric_df = df.select_dtypes(include=np.number)

print("Highest Mean Feature:")
print(numeric_df.mean().idxmax())

print("\nHighest Standard Deviation Feature:")
print(numeric_df.std().idxmax())
```

```python
# Q4. Distribution Analysis

columns = [
    "rainfall_mm",
    "temperature_c",
    "fertilizer_kg",
    "yield_ton_per_hectare"
]

for col in columns:
    plt.figure(figsize=(6,4))
    plt.hist(df[col], bins=20)

    plt.title(col)
    plt.xlabel(col)
    plt.ylabel("Frequency")

    plt.show()
```

```python
# Q5. Crop Type Analysis

print(df["crop_type"].value_counts())

sns.countplot(data=df, x="crop_type")
plt.xticks(rotation=45)
plt.show()

print("Most Frequent Crop:")
print(df["crop_type"].mode()[0])
```

```python
# Q6. Soil Type Analysis

print(df["soil_type"].value_counts())

sns.countplot(data=df, x="soil_type")
plt.xticks(rotation=45)
plt.show()

print("Most Common Soil:")
print(df["soil_type"].mode()[0])
```

```python
# Q7. Yield Distribution

plt.figure(figsize=(6,4))

plt.hist(df["yield_ton_per_hectare"], bins=20)

plt.title("Yield Distribution")
plt.xlabel("Yield")
plt.ylabel("Frequency")

plt.show()
```

```python
# Q8. Scatter Plot Analysis

plt.figure(figsize=(6,4))

plt.scatter(
    df["rainfall_mm"],
    df["yield_ton_per_hectare"]
)

plt.xlabel("Rainfall")
plt.ylabel("Yield")

plt.show()

plt.figure(figsize=(6,4))

plt.scatter(
    df["fertilizer_kg"],
    df["yield_ton_per_hectare"]
)

plt.xlabel("Fertilizer")
plt.ylabel("Yield")

plt.show()
```

```python
# Q8. Scatter Plot Analysis

plt.figure(figsize=(6,4))

plt.scatter(
    df["rainfall_mm"],
    df["yield_ton_per_hectare"]
)

plt.xlabel("Rainfall")
plt.ylabel("Yield")

plt.show()

plt.figure(figsize=(6,4))

plt.scatter(
    df["fertilizer_kg"],
    df["yield_ton_per_hectare"]
)

plt.xlabel("Fertilizer")
plt.ylabel("Yield")

plt.show()
```

```python
# Q10. Group-Based Analysis

print("Average Yield by Crop Type")

print(
    df.groupby("crop_type")["yield_ton_per_hectare"].mean()
)

print("\nAverage Yield by Soil Type")

print(
    df.groupby("soil_type")["yield_ton_per_hectare"].mean()
)
```

```python
# Q11. Feature Encoding

categorical_columns = df.select_dtypes(include="object").columns

print(categorical_columns)

encoded_df = pd.get_dummies(
    df,
    columns=categorical_columns
)

display(encoded_df.head())
```

```python
# Q12. Feature Selection

X = encoded_df.drop(
    "yield_ton_per_hectare",
    axis=1
)

y = encoded_df["yield_ton_per_hectare"]

print("Target Variable: yield_ton_per_hectare")
```

```python
# Q13. Train-Test Split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("X_train Shape:", X_train.shape)
print("X_test Shape:", X_test.shape)
print("y_train Shape:", y_train.shape)
print("y_test Shape:", y_test.shape)
```

```python
# Q14. Linear Regression Model

from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(
    X_train,
    y_train
)

print("Intercept:")

print(model.intercept_)

coef_df = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": model.coef_
})

display(coef_df)

print("Highest Positive Coefficient:")

display(
    coef_df.sort_values(
        by="Coefficient",
        ascending=False
    ).head(1)
)
```
